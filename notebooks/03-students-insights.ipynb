```python
import pandas as pd
```

```python
file_path = '../data/processed/engineered_data.csv'
df = pd.read_csv(file_path) ; df
```

```python
# Find the top 10 students
top_students = df.sort_values(by='gpa', ascending=False)
```

```python
top_students.head()
```

```python
top_students[['student_id', 'full_name', 'gender', 'gpa']].to_csv('../data/processed/top_students.csv', index=False)
```

```python
at_risk = df[df['risk_score'] > 0]
```

```python
at_risk.head()
```

```python
at_risk[['student_id', 'full_name', 'gpa', 'risk_score', 'engagement_level']].to_csv('../data/processed/at_risk_students.csv', index=False)
```

```python
engagement_counts = df['engagement_level'].value_counts()
engagement_counts.to_csv('../data/processed/engagement_counts.csv', header=True)
```
